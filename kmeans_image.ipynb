```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
x=np.array(Image.open('im.jpg').convert("L"))
plt.imshow(x,plt.cm.gray)
plt.show()
```

```python
m=np.max(x)+1
intensity=[]
for i in range(m):
    value=np.sum(x==i)
    intensity.append(value)
```

```python
plt.plot(intensity)
plt.xlim([0,255])
plt.grid()
plt.show()
```

```python
k=3
mu=np.arange(0,k)*m/(k+1)
mu
```

```python
intensity=np.array(intensity)
intensity_count=np.argwhere(intensity)
index=len(intensity_count)
labels=np.zeros(m)
labels.shape,index
```

```python
for it in range(5):
    for i in range(index):
        distance=np.sqrt((intensity_count[i]-mu)**2)
        value=np.argwhere(distance==np.min(distance))
        labels[intensity_count[i]]=value[0,0]
    
    for i in range(k):
        r=labels==i
        mu[i]=np.sum(intensity[r])/np.sum(r)
    
    
    
    
    
```

```python
[m,n]=x.shape
new_label=np.zeros([m,n])
for i in range(m):
    for j in range(n):
        index=np.abs(x[i,j]-mu)
        center=np.argwhere(np.min(index)==index)
        new_label[i,j]=center[0,0]
        
        
```

```python
plt.imshow(new_label)
```
